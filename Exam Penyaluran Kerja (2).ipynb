```python
import pandas as pd
import os

# Load semua file CSV
folder_path = './Exam/'

customer_info = pd.read_csv(os.path.join(folder_path, 'Customer_Info.csv'))
location_data = pd.read_csv(os.path.join(folder_path, 'Location_Data.csv'))
payment_info = pd.read_csv(os.path.join(folder_path, 'Payment_Info.csv'))
service_options = pd.read_csv(os.path.join(folder_path, 'Service_Options.csv'))
online_services = pd.read_csv(os.path.join(folder_path, 'Online_Services.csv'))
status_analysis = pd.read_csv(os.path.join(folder_path, 'Status_Analysis.csv'))

# Gabungkan semua dataframe berdasarkan customer_id
df = customer_info.copy()

# List dataframe untuk digabungkan
dataframes = [
    location_data,
    payment_info,
    service_options,
    online_services,
    status_analysis
]

# Gabungkan semua
for dataframe in dataframes:
    df = pd.merge(df, dataframe, on='customer_id', how='left')

# Hapus kolom duplikat jika ada
df = df.loc[:, ~df.columns.duplicated()]

print(f"Dataframe gabungan berhasil dibuat:")
print(f"Shape: {df.shape}")
print(f"Kolom: {len(df.columns)}")
print(f"Baris: {len(df)}")

# Tampilkan beberapa baris pertama
print("\nPreview:")
df.head()
```
