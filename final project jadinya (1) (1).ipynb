```python
#1. import library
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# STEP 1 – Fix data types
df1['Customer_ID'] = df1['Customer_ID'].astype('Int64')
df2['Customer_ID'] = df2['Customer_ID'].astype('Int64')
df2['Product_Id'] = df2['Product_Id'].astype('Int64')
df3['Product_Id'] = df3['Product_Id'].astype('Int64')

# STEP 2 – Drop duplicates
df1 = df1.drop_duplicates(subset=['Customer_ID'])
df2 = df2.drop_duplicates(subset=['Transaction_ID'])
df3 = df3.drop_duplicates(subset=['Product_Id'])

# STEP 3 – Select needed columns from df3
df3_small = df3[['Product_Id', 'Product_Category', 'Product_Type', 'Products']]

# STEP 4 – JOIN (lighter than merge)
df1 = df1.set_index('Customer_ID')
df2 = df2.set_index('Customer_ID')

merged1 = df1.join(df2, how='left')

merged1 = merged1.reset_index().set_index('Product_Id')
df3_small = df3_small.set_index('Product_Id')

final_df = merged1.join(df3_small, how='left').reset_index()

df = final_df
df.head()
```

```python
# Deteksi outlier dengan IQR
# Hitung Q1, Q3, dan IQR untuk kolom numerik
Q1 = df[numerik].quantile(0.25)
Q3 = df[numerik].quantile(0.75)
IQR = Q3 - Q1
#Tentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Deteksi outlier
df_outlier = df[((df[numerik] < lower_bound) | (df[numerik] > upper_bound)).any(axis=1)]
print("Jumlah baris outlier:", df_outlier.shape[0])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Future engineering
df['Order_Year'] = df['Date'].dt.year
df['Order_Month'] = df['Date'].dt.month
df['Order_Day'] = df['Date'].dt.day
df['Order_Weekday'] = df['Date'].dt.weekday  # 0=Monday, 6=Sunday
df['Order_WeekOfYear'] = df['Date'].dt.isocalendar().week
df['Order_Quarter'] = df['Date'].dt.quarter
```

```python
print(df['Payment_Method'].value_counts())
```

```python
df['Days_Since_First_Purchase'] = (df['Date'] - df.groupby('Customer_ID')['Date'].transform('min')).dt.days
df['Days_Since_Last_Purchase'] = (df.groupby('Customer_ID')['Date'].transform('max') - df['Date']).dt.days
df[['Days_Since_First_Purchase', 'Days_Since_Last_Purchase']].head()
```

```python
df['Amount_Category'] = pd.qcut(df['Amount'], q=4,
                               labels=['Low_Spend', 'Medium_Spend', 'High_Spend', 'VIP_Spend'])
df[['Amount_Category','Amount']].head()
```
