```python
#4.Missing value
#menghapus missing value
#jalankan jika ada missing value, jika tidak, tidak lanjut ke tahap berikutnya
df["company"]  = df["company"].fillna(0)
df["agent"]    = df["agent"].fillna(0)
df["country"]  = df["country"].fillna("Unknown")
df["children"] = df["children"].fillna(0)
```

```python
#5. mengkonfersi type data (order date, time order, dan time order picked)
#konversi kolom tanggal
df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"], errors="coerce")

df["arrival_date"] = pd.to_datetime(
    df["arrival_date_year"].astype(str) + "-" +
    df["arrival_date_month"].astype(str) + "-" +
    df["arrival_date_day_of_month"].astype(str),
    errors="coerce"
)

df.head()
```

```python
#5. Drop kolom yang tidak digunakan pada EDA
df = df.drop(columns=[
    'bookingID',
    'reservation_status',
    'reservation_status_date',
    'arrival_date_year',
    'arrival_date_day_of_month'
])
```

```python
#Mengelompokan setiap kolom berdasarkan tipenya
#memisahkan data numerik dan kategorik
numerik = df.select_dtypes(include=['number']).columns.tolist()
kategorik = df.select_dtypes(exclude=['number']).columns.tolist()

df[numerik].head()
```

```python
#6. Feature Engineering (buat kolom baru)
df["total_nights"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]
df["total_guests"] = df["adults"] + df["children"] + df["babies"]
df["has_special_requests"] = (df["total_of_special_requests"] > 0).astype(int)

df.head()
```

```python
#6. Deteksi outlier dengan IQR
# Hitung Q1, Q3, dan IQR untuk kolom numerik
# Tentukan kolom numerik yang mau dicek outlier
outlier_cols = ["lead_time", "adr", "total_nights"]

Q1 = df[outlier_cols].quantile(0.25)
Q3 = df[outlier_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_outlier = df[
    ((df[outlier_cols] < lower_bound) | (df[outlier_cols] > upper_bound)).any(axis=1)
]

print("Outlier:", df_outlier.shape[0])
```

```python
#6. Outlier
#Hapus baris yang memiliki outlier di kolom numerik
df = df[
    ~((df[outlier_cols] < lower_bound) | (df[outlier_cols] > upper_bound)).any(axis=1)
]

print("Total data", df.shape[0])
```
